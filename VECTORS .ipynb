```python
a = np.array([2, 3])

b = np.array([4, 4])

a_dot_b = np.dot(a, b)

print(a_dot_b)
```

```python
a = np.array([2, 3])

b = np.array([4, 4])

a_cross_b = np.cross(a, b)
print(a_cross_b)
     
```
